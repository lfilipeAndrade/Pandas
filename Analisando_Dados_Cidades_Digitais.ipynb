```python
import pandas as pd
import numpy as np
```

```python
cidadesDigitais = pd.read_csv('cidades_digitais.csv',sep=';',encoding='latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.count()
```

```python
# Excluir a coluna 'IBGE', pois a mesma representa apenas o ID da cidade no IBGE e não será utilizada na análise
del cidadesDigitais['IBGE']
```

```python
# Agrupando o df pela coluna 'UF' usando o group by
cidadeUF_1 = cidadesDigitais.groupby('UF')
```

```python
# Quantidade de grupos do df
len(cidadeUF_1.groups)
```

```python
# Definir uma indexação por 'STATUS'
cidadeUF_2 = cidadesDigitais.set_index('STATUS')
```

```python
# Aplicando o group by 
cidadeUF_2 = cidadeUF_2.groupby(level='STATUS')
```

```python
cidadeUF_2.head()
```

```python
# Definindo 2 indices para o df
cidadeUF_3 = cidadesDigitais.set_index(['UF','STATUS'])
```

```python
cidadeUF_3
```

```python
cidadeUF_3 = cidadeUF_3.groupby(['UF','STATUS'])
```

```python
cidadeUF_3.sum()
```

```python
cidadeUF_3.aggregate(np.sum)
```

```python
cidadeUF_3.aggregate([np.sum, np.mean, np.size])
```

```python
cidadesUF_4 = cidadesDigitais.set_index([ 'STATUS', 'CIDADE', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO', 'POPULAÇÃO', 'PONTOS_ATENDIDOS'])
```

```python
cidadesUF_4 = cidadesUF_4.groupby(level = ['STATUS', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO'])
```

```python
cidadesUF_4.sum()
```
